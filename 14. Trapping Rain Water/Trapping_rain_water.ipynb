Trapping Rain Water
You are given an array of non-negative integers height which represent an elevation map. Each value height[i] represents the height of a bar, which has a width of 1.

Return the maximum area of water that can be trapped between the bars.



```python
def trappingRainWater(height: list[int]):
    if not height:
        return 0

    
    lPointer, rPointer = 0, len(height) - 1
    maxLeft = height[lPointer]
    maxRight = height[rPointer]
    res = 0
    while lPointer < rPointer:
        if maxLeft < maxRight: 
            lPointer +=1
            maxLeft = max(maxLeft, height[lPointer])
            res += maxLeft - height[lPointer]
        else:
            rPointer -= 1
            maxRight = max(maxRight, height[rPointer])
            res += maxRight - height[rPointer]
    return res
        
```

```python
trappingRainWater([0,2,0,3,1,0,1,3,2,1])
```
